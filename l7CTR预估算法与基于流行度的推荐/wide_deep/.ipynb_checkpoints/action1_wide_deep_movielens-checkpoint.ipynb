```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from deepctr.inputs import SparseFeat,get_feature_names
from deepctr.models import WDL,FNN,NFM,PNN,AFM
```

```python
# 数据加载
data = pd.read_csv('movielens_sample.txt')
sparse_features = ['movie_id','user_id','gender','age','occupation','zip']
target = ['rating']
```

```python

# 对特征标签进行编码
for feature in sparse_features:
    lbe = LabelEncoder()
    data[feature] = lbe.fit_transform(data[feature])
# 计算每个特征中的 不同特征值的个数
fixlen_feature_columns = [SparseFeat(feature,data[feature].nunique()) for feature in sparse_features]
linear_feature_columns = fixlen_feature_columns
dnn_feature_columns = fixlen_feature_columns
feature_name = get_feature_names(linear_feature_columns + dnn_feature_columns)

# 将数据集切分为训练集和测试集
train, test = train_test_split(data,test_size=0.2)
train_model_input = {name:train[name].values for name in feature_name}
test_model_input = {name:test[name].values for name in feature_name}
```

```python

# 使用WDL进行训练
model = WDL(linear_feature_columns,dnn_feature_columns,task='regression')
model.compile('adam','mse',metrics=['mse'],)
history = model.fit(train_model_input,train[target].values,batch_size=256,epochs=100,verbose=True,validation_split=0.2,)
# 使用WDL进行预测
pred_ans = model.predict(test_model_input,batch_size=256)
# 输出RMSE或者MSE
mse = round(mean_squared_error(test[target].values,pred_ans),4)
rmse = mse ** 0.5
print('使用WDL进行预测,test RMSE',rmse)
```

```python
# 使用FNN进行训练
fnn_model = FNN(linear_feature_columns,dnn_feature_columns,task='regression')
fnn_model.compile('adam','mse',metrics=['mse'],)
history = fnn_model.fit(train_model_input,train[target].values,batch_size=256,epochs=100,verbose=True,validation_split=0.2)
# 使用FNN进行预测
pred_ans = fnn_model.predict(test_model_input,batch_size=256)
# 输出RMSE或者MSE
mse = round(mean_squared_error(test[target].values,pred_ans),4)
rmse = mse ** 0.5
print('使用FNN进行预测,test RMSE',rmse)
```

```python
# 使用NFM进行训练
nfm_model = NFM(linear_feature_columns,dnn_feature_columns,task='regression')
nfm_model.compile('adam','mse',metrics=['mse'],)
history = nfm_model.fit(train_model_input,train[target].values,batch_size=256,epochs=100,verbose=True,validation_split=0.2)
# 使用NFM进行预测
pred_ans = nfm_model.predict(test_model_input,batch_size=256)
# 输出RMSE或者MSE
mse = round(mean_squared_error(test[target].values,pred_ans),4)
rmse = mse ** 0.5
print('使用NFM进行预测,test RMSE',rmse)
```
